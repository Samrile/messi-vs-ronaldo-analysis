```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mdata = pd.read_csv(r"C:\Users\samkh\Downloads\Lionel_Messi_Detailed_Career_Stats.csv")
```

```python
rdata = pd.read_csv(r"C:\Users\samkh\Downloads\Cristiano_Ronaldo_Career_Stats.csv")
```

```python
mdata
```

```python
mdata = mdata.drop_duplicates()
```

```python
rdata
```

```python
mdata["Player"] = "Lionel Messi"
rdata["Player"] = "Cristiano Ronaldo"
```

```python
mdata
```

```python
rdata
```

```python
combo_data = pd.concat([mdata,rdata])
```

```python
combo_data
```

```python
comparison_data = combo_data[['Year', 'Player', 'Goals Scored', 'Assists']]
```

```python
comparison_data
```

```python
years = comparison_data['Year'].unique()
```

```python
years
```

```python
mgoals = []
rgoals = []
```

```python
for year in years:
    mgoals.append(comparison_data[(comparison_data["Year"] == year) & (comparison_data["Player"] == "Lionel Messi")]["Goals Scored"].values[0])
    rgoals.append(comparison_data[(comparison_data["Year"] == year) & (comparison_data["Player"] == "Cristiano Ronaldo")]["Goals Scored"].values[0])
```

```python
x = np.arange(len(years))
width = 0.35
```

```python
fig, ax = plt.subplots(figsize=(12, 6))
```

```python
fig, ax = plt.subplots(figsize=(12, 12))
bars1 = ax.bar(x - width/2, mgoals, width, label='Lionel Messi', color='blue')
bars2 = ax.bar(x + width/2, rgoals, width, label='Cristiano Ronaldo', color='red')
ax.set_xlabel("Years",fontsize=12)
ax.set_ylabel("Goals Scored",fontsize=12)
ax.set_title("Goals Scored",fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(years, rotation=45)
ax.legend()
plt.tight_layout()
plt.show()
```
